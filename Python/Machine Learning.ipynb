```python
# Machine Learning via Python!
import sklearn
# There are two types of machine learning, 
# When output is catagorial, or continous. 
# In this example we look at a continious predictor...
```

```python
# A function that creates fake data that has 3 predictors for a result
def make_data(N,sigma=0.1):
    predictie = np.linspace(0,2*np.pi,N)
    predictor = np.zeros((N,3))
    predictor[:,0] = np.sin(predictie) + np.random.normal(loc=0,scale=sigma,size=N)
    predictor[:,1] = np.cos(predictie) + np.random.normal(loc=0,scale=sigma,size=N)
    predictor[:,2] = np.exp(-predictie/3) + np.random.normal(loc=0,scale=sigma,size=N)
    return predictie,predictor
# Lets use our function now to create some data that we can train the algorithm on
training_out,training_inp = make_data(2000,0.1)
plt.xlabel('Predictie')
plt.ylabel('Predictor Values')
plt.plot(training_out,training_inp)
plt.legend(('0','1','2'),title='Predictors')
```

```python
# Import the machine learning algorithm, this is one of several available
from sklearn import tree
# Create the machine learning object
dtr = tree.DecisionTreeRegressor()
# Train the algorithm on the training data
dtr.fit(training_inp, training_out)
```

```python
# Now we want to test the algorithm on another set of data, so generate another set, use a larger noise though...
test_out,test_inp = make_data(1000,0.2)
# This tells us how well the algorithm is doing on this data
dtr.score(test_inp, test_out)
```

```python
# Of course we dont always know the result, thats sorta the point afterall...
# A trained machine learning object shoulb be able to precit the result based only on the predictors.
dtr.predict(np.array((.5,.5,.3)))
```

```python
N = 5000
atest = np.zeros((N,3))
atest[:,0] = np.random.uniform(-1,1,N)
atest[:,1] = np.random.uniform(-1,1,N)
atest[:,2] = np.random.uniform(0,1,N)
out = dtr.predict(atest)
```

```python
plt.plot(out,atest[:,0],'.')
plt.plot(out,atest[:,1],'.')
plt.plot(out,atest[:,2],'.')
```
